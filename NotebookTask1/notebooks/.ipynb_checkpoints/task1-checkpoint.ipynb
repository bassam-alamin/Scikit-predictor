Find the top 10 countries that grew "Gross domestic product per capita" the most over the last
decade.

```python
import pandas as pd

# Load the dataset
we_url = "./NotebookTask1/we_data/WEOOct2020all.xlsx"

we_df = pd.read_excel(we_url)

# Select the relevant columns
columns_to_keep = ['Country', 'Subject Descriptor', 2013, 2014, 2015, 2016, 2017, 2018, 2019, 2020, 2021, 2022]
we_df = we_df[columns_to_keep]

# Filter only rows with Gross domestic product per capita in the subject descriptor
gdp_per_capita_df = we_df[we_df['Subject Descriptor'] == 'Gross domestic product per capita']

# Calculate the growth rate of GDP per capita for each country

gdp_per_capita_df['Growth Rate'] = ((gdp_per_capita_df[2022] - gdp_per_capita_df[2013]) / gdp_per_capita_df[2013]) * 100

# Sort the countries based on the growth rate in descending order
gdp_per_capita_df = gdp_per_capita_df.sort_values(by='Growth Rate', ascending=False)

# Select the top 10 countries with the highest growth rates
top_10_countries = gdp_per_capita_df.head(10)

# Display the top 10 countries
print(top_10_countries[['Country','Subject Descriptor', 'Growth Rate']])
# Todo: Get the name of country using the WEO country code 
# Todo: Get rid of the settingscopywarning
# Todo: use classes or functions for this snippet.
```

Draw OECD countries' "Population" growth over the last decade.

```python
import matplotlib.pyplot as plt

# Assuming you have a DataFrame named 'countries_dataframe' containing the data

# Step 1: Filter the OECD countries
oecd_countries = ['Austria', 'Belgium', 'Chile', 'Czech Republic', 'Denmark', 'Estonia', 'Finland',
                  'France', 'Germany', 'Greece', 'Hungary', 'Iceland', 'Ireland', 'Israel', 'Italy',
                  'Japan', 'Korea', 'Latvia', 'Lithuania', 'Luxembourg', 'Mexico', 'Netherlands',
                  'New Zealand', 'Norway', 'Poland', 'Portugal', 'Slovak Republic', 'Slovenia',
                  'Spain', 'Sweden', 'Switzerland', 'Turkey', 'United Kingdom', 'United States']

oecd_population_data = we_df[(we_df['Country'].isin(oecd_countries)) &
                                           (we_df['Subject Descriptor'] == 'Population')]

# Step 2: Extract population data for the last decade (2013 to 2022)
years = [int(2013 + i) for i in range(10)]
oecd_population_growth = oecd_population_data[years]

# Step 3: Plot the population growth of OECD countries over the last decade
plt.figure(figsize=(10, 6))
for country in oecd_countries:
    plt.plot(years, oecd_population_growth[oecd_population_data['Country'] == country].values[0], label=country)

plt.xlabel('Year')
plt.ylabel('Population')
plt.title('Population Growth of OECD Countries (2013-2022)')
plt.legend()
plt.grid(True)
plt.show()

# Todo: use classes or functions for this snippet.
```

Create 5 clusters out of the countries using GDP and "Volume of exports of goods".
a. Draw the charts (x-axis - GDP, y - volume).
b. Add labels for the top 5 countries according to the GDP on the dots representing countries
in each cluster.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
# Set the backend to TkAgg
import matplotlib
matplotlib.use("nbAgg")


# Assuming you have the data in a DataFrame named 'data'
# Extract the GDP and Volume of exports of goods columns
gdp_data = we_df[we_df['Subject Descriptor'] == 'Gross domestic product']
volume_exports_data = we_df[we_df['Subject Descriptor'] == 'Volume of exports of goods']
new_we_df = we_df.drop_duplicates(subset="Country")

def sum_fields_2020(country, subject_descriptor):
    gdp_column = we_df[we_df['Country'] == country]
    gdp_column = gdp_column[we_df["Subject Descriptor"] == subject_descriptor][2019]
    gdp_numeric = pd.to_numeric(gdp_column, errors='coerce').fillna(0).astype(int)
    total = gdp_numeric.sum()
    return total

# Create new columns 'gdp' and 'volume' in the DataFrame and calculate the sums
for i, country in new_we_df.iterrows():
    new_we_df.loc[i, 'gdp'] = sum_fields_2020(country["Country"], "Gross domestic product")
    new_we_df.loc[i, 'volume'] = sum_fields_2020(country["Country"], "Volume of exports of goods")
    
    
print(len(new_we_df["gdp"]))
print(len(new_we_df["volume"]))
print(new_we_df[["gdp", "volume"]].head(5))
# Todo: To activate below code

# Create a feature matrix
feature_matrix = np.column_stack((new_we_df["gdp"], new_we_df["volume"]))

# Standardize the data
scaler = StandardScaler()
scaled_features = scaler.fit_transform(feature_matrix)

# Apply K-means clustering to create 5 clusters
kmeans = KMeans(n_clusters=5, random_state=42)
clusters = kmeans.fit_predict(scaled_features)

# Add the cluster labels to the DataFrame
new_we_df['Cluster'] = clusters
# Find the top 5 countries with the highest GDP values
# top_5_countries = new_we_df.nlargest(5, 2019, keep='first')

# Plot the clusters and label the top 5 countries
plt.figure(figsize=(10, 6))
for cluster_num in range(5):
    cluster_data = new_we_df[new_we_df['Cluster'] == cluster_num]
    plt.scatter(cluster_data["gdp"], cluster_data['volume'], label=f'Cluster {cluster_num + 1}')

    # Add labels for the top 5 countries with the highest GDP values
    for i, country in cluster_data.sort_values(by='gdp', ascending=False).head(5).iterrows():
        plt.annotate(country['Country'], xy=(country["gdp"], country['volume']),
                     xytext=(5, 5), textcoords='offset points', ha='left', va='bottom')

plt.xlabel('GDP (current prices)')
plt.ylabel('Volume of exports of goods')
plt.title('Clusters of Countries based on GDP and Volume of Exports')
plt.legend()
plt.grid(True)
plt.show()

```

Create a predictor (use scikit) to predict GDP per capita (exclude other GDP-related fields).

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error

we_df.columns = we_df.columns.astype(str)
we_df = we_df[we_df["Subject Descriptor"] == "Gross domestic product per capita"].dropna()

# Split the data into training and testing sets
training_set, testing_set = train_test_split(
    we_df[["2015", "2016", "2017", "2018", "2019", "2020"]], 
    test_size=0.25)

print(training_set.shape)
training_test = training_set.iloc[:,1]
print(training_test.shape)



# Create the predictor and train it
regressor = LinearRegression()
regressor.fit(training_set, training_test)

# Make predictions on the training and testing sets
y_train_pred = regressor.predict(training_set)
y_test_pred = regressor.predict(testing_set)
print(y_train_pred.shape)

# Calculate the mean squared error (MSE) on the training and testing data sets
# mse_train = mean_squared_error(training_set, y_train_pred)
# mse_test = mean_squared_error(training_test.iloc[:, 0], y_test_pred)
# 
# print(f"Mean Squared Error on Training Set: {mse_train}")
# print(f"Mean Squared Error on Testing Set: {mse_test}")

# Name the fields that were used during training
print("Fields used during training:", ['Scale', '2010'])

# Find the top 5 fields/features that contribute the most to the predictions
# We'll use feature importance if we were using a different model like Random Forest
# For Linear Regression, we can use the coefficients of the features
print(regressor.coef_)
top_features = pd.DataFrame({'Feature': ["2015", "2016", "2017", "2018", "2019", "2020"], 'Coefficient': regressor.coef_})
top_features = top_features.nlargest(5, 'Coefficient')

print("Top 5 fields contributing the most to predictions:")
print(top_features)
```

```python
# Select the top 5 features and target
# Split the data into training and testing sets
training_set, testing_set = train_test_split(
    top_features, 
    test_size=0.25)

print(training_set.shape)
training_test = training_set.iloc[:,1]
print(training_test.shape)



# Create the predictor and train it
regressor_top = LinearRegression()
regressor_top.fit(training_set, training_test)

# Make predictions on the training and testing sets
y_train_pred = regressor_top.predict(training_set)
y_test_pred = regressor_top.predict(testing_set)
print(y_train_pred.shape)

# Calculate the mean squared error (MSE) on the training and testing data sets
# mse_train = mean_squared_error(training_set, y_train_pred)
# mse_test = mean_squared_error(training_test.iloc[:, 0], y_test_pred)
# 
# print(f"Mean Squared Error on Training Set: {mse_train}")
# print(f"Mean Squared Error on Testing Set: {mse_test}")

# Name the fields that were used during training
print("Fields used during training:", ['Scale', '2010'])

# Find the top 5 fields/features that contribute the most to the predictions
# We'll use feature importance if we were using a different model like Random Forest
# For Linear Regression, we can use the coefficients of the features
print(regressor_top.coef_)
```
